```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from typing import List
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
def load_all_csvs(csv_paths: List[str]) -> pd.DataFrame:
        """Load and combine multiple experiment CSVs."""
        dfs = []
        for path in csv_paths:
            df = pd.read_csv(path, index_col=0)
            df['source_file'] = Path(path).parent.name  # track which experiment it came from
            dfs.append(df)

            combined = pd.concat(dfs, ignore_index=True)
                
        return combined
```

```python
paths = ['data/results/2428981015/ni_n_sweep/ni10n_10000_red3_1_sweep//results_sd2428981015.csv', 'data/results/2428981015/ni_n_sweep/ni10n_20000_red3_1_sweep//results_sd2428981015.csv', 'data/results/2428981015/ni_n_sweep/ni10n_40000_red3_1_sweep//results_sd2428981015.csv', 'data/results/2428981015/ni_n_sweep/ni10n_60000_red3_1_sweep//results_sd2428981015.csv']['data/results/2428981015/ni_n_sweep/ni10n_10000_red3_1_sweep//results_sd2428981015.csv', 'data/results/2428981015/ni_n_sweep/ni10n_20000_red3_1_sweep//results_sd2428981015.csv', 'data/results/2428981015/ni_n_sweep/ni10n_40000_red3_1_sweep//results_sd2428981015.csv', 'data/results/2428981015/ni_n_sweep/ni10n_60000_red3_1_sweep//results_sd2428981015.csv']
df = load_all_csvs(paths)
df['gap_size_range_tuple'] = df['gap_size_range'].apply(ast.literal_eval)
df['reduce_triggerSz_sizeLim_tuple'] = df['reduce_triggerSz_sizeLim'].apply(ast.literal_eval)
df
```
